```python
import json
import os

# Load climate zones data
with open('../data/BuildingClimateZones_CEC_2015_-254197201195074155.geojson') as f:
    climate_data = json.load(f)

print('Type:', climate_data.get('type'))
print('Features count:', len(climate_data.get('features', [])))
if climate_data.get('features'):
    print('\nFirst feature:')
    print('Keys:', list(climate_data['features'][0].keys()))
    print('Geometry type:', climate_data['features'][0]['geometry']['type'])
    print('Properties:', climate_data['features'][0].get('properties'))
    print('\nCoordinates sample:', str(climate_data['features'][0]['geometry']['coordinates'])[:200])
```

```python
# Load MDHD data
with open('../data/MDHD_Dashboard_ArcGIS_Updated_Nov_6901160749949423404.geojson') as f:
    mdhd_data = json.load(f)

print('Type:', mdhd_data.get('type'))
print('Features count:', len(mdhd_data.get('features', [])))
if mdhd_data.get('features'):
    print('\nFirst feature:')
    print('Keys:', list(mdhd_data['features'][0].keys()))
    print('Geometry type:', mdhd_data['features'][0]['geometry']['type'])
    print('Properties:', mdhd_data['features'][0].get('properties'))
    print('\nCoordinates sample:', str(mdhd_data['features'][0]['geometry']['coordinates'])[:200])
```

```python
# Get bounding box
def get_bounds(features):
    lons = []
    lats = []
    for f in features[:10]:
        coords = f['geometry']['coordinates']
        if f['geometry']['type'] == 'Polygon':
            for ring in coords:
                for lon, lat in ring:
                    lons.append(lon)
                    lats.append(lat)
        elif f['geometry']['type'] == 'MultiPolygon':
            for poly in coords:
                for ring in poly:
                    for lon, lat in ring:
                        lons.append(lon)
                        lats.append(lat)
    return min(lons), min(lats), max(lons), max(lats)

bounds = get_bounds(climate_data['features'])
print('Bounds (west, south, east, north):', bounds)
print('Center:', [(bounds[0]+bounds[2])/2, (bounds[1]+bounds[3])/2])
```
